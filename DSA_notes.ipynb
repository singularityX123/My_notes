```c++17
/*时间复杂度  空间复杂度  是否稳定*/

// 简单选择排序  O(n^2)  O(1)  不稳定    
// 每次从未排序部分选择最小元素放到已排序部分的末尾
void selectionSort(int arr[], int n) {
    for (int i = 0; i < n-1; i++) {
        int min_idx = i;
        for (int j = i+1; j < n; j++) {
            if (arr[j] < arr[min_idx]) {
                min_idx = j;
            }
        }
        // 交换
        int temp = arr[i];
        arr[i] = arr[min_idx];
        arr[min_idx] = temp;
    }
}

// 冒泡排序  O(n^2)  O(1)  稳定    
// 重复遍历数组，每次比较相邻元素，如果顺序错误就交换
void bubbleSort(int arr[], int n) {
    for (int i = 0; i < n-1; i++) {
        for (int j = 0; j < n-i-1; j++) {
            if (arr[j] > arr[j+1]) {
                // 交换
                int temp = arr[j];
                arr[j] = arr[j+1];
                arr[j+1] = temp;
            }
        }
    }
}
```
