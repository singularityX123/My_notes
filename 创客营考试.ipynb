快速排序

```c++17
void qsort(int arr[], int left, int right) {
    if (left >= right) {
        return;
    }

    int pivot = arr[(left + right) / 2];
    int i = left;
    int j = right;
    while (i <= j) {
        while (arr[i] < pivot) {
            i++;
        }
        while (arr[j] > pivot) {
            j--;
        }
        if (i <= j) {
            int temp = arr[i];
            arr[i] = arr[j];
            arr[j] = temp;
            i++;
            j--;
        }
    }
    qsort(arr, left, j);
    qsort(arr, i, right);
}



int arr[] = {9, 8, 7, 6, 5, 4, 7, 3, 2, 1, 0};
qsort(arr, 0, 10);

for (int i = 0; i < 11; i++) {
    printf("%d ", arr[i]);
}


```

单向链表

```c++17
typedef int data_t;
typedef struct Node {
    data_t data;
    struct Node* next;
} listnode, *linklist;

linklist list_create(){
    linklist H = (linklist)malloc(sizeof(listnode)); 

    if (H == NULL){
        return NULL;
    }

    H->data = 0; 
    H->next = NULL; 
    
    return H;
};
```
