#### 概念

```c++17
一般指封装了系统调用的应用编程接口函数，与硬件有强耦合关系（每次调用都需要切换到内核态，引起系统调用）
```

#### 文件描述符

```c++17
英文缩写(File Descriptor) fd
为非负整数，0开始分配，依次递增（linux中默认一个进程可以打开的数量是0-1023，1024个）
文件IO操作通过文件描述符完成

规定:0为标准输入，1为标准输出，2为标准错误输出   人为每次打开文件从3开始分配
```

#### umask设置权限 ulimit观察打开数量

```c++17
umask：用来设置作为默认文件或目录的初始权限的掩码
    e.g. umask 002
        即原先权限按位异或002（0-010）

ulimit：用来设置用户可以同时打开的文件描述符的最大数量
    e.g. ulimit -n
        即查看用户可以同时打开的文件描述符的最大数量
        ulimit -n 1024
        即设置用户可以同时打开的文件描述符的最大数量为1024
        
```
