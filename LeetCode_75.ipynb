<big>1.交替合并字符串</big>

```c++17
char word1[] = "ace";  
char word2[] = "bdfgh"; 
```

<small>双指针</samll>

```c++17
char* mergeAlternately(char* word1, char* word2) {
    int len1 = strlen(word1);
    int len2 = strlen(word2);
    
    char* result = (char*)malloc(len1 + len2 + 1);
    if (!result) return NULL;
    
    int i = 0, j = 0, k = 0;
    
    while (i < len1 || j < len2) {
        if (i < len1) result[k++] = word1[i++];
        if (j < len2) result[k++] = word2[j++];
    }
    
    result[k] = '\0';
    return result;
}
```

```c++17
char* result = mergeAlternately(word1, word2);
printf("%s\n", result);
free(result);//谁用谁负责释放内存
```

<big>2.字符串的最大公因子</big>

```c++17
char word1[] = "ababababab";  
char word2[] = "abab"; 
```

```c++17
#include <string.h>
#include <stdlib.h>
#include <stdio.h>

//欧几里得迭代法
/*int gcd(int a, int b) {
    while (b != 0) {
        int temp = b;
        b = a % b;
        a = temp;
    }
    return a;
}*/

//欧几里得递归法
int gcd(int a, int b) {
    if (b == 0) {
        return a;
    }
    return gcd(b, a % b);
}

char* gcdOfStrings(char* str1, char* str2) {
    //如果两个字符串逆序拼接不相等，则没有最大公因子
    if (strcmp(strcat(str1, str2), strcat(str2, str1)) != 0) {
        static const char empty[] = "";
        return empty;

    }else{
        //辗转相除法（欧几里得算法）求最大公约字符长度gcd_len
        //取第一个字符串前gcd_len个字符即为最大公约字符串
        int gcd_len = gcd(strlen(str1), strlen(str2));
        char* result = (char*)malloc(gcd_len + 1);
        strncpy(result, str1, gcd_len);
        result[gcd_len] = '\0';

        return result;

    }
}
```

```c++17
printf("%s", gcdOfStrings(word1, word2));
free(result);
```
