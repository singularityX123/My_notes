取余  取模

// 取余(remainder)和取模(modulo)都是计算除法后的余数，但在处理负数时有所不同
// 取余： 被除数 - (除数 * 商)          其中商向0取整 
// 取模： (被除数 % 除数 + 除数) % 除数


// 取余结果的符号与被除数相同
int result = -7 % 4;  // 结果是 -3
printf("result = %d\n", result); // 表示除法后的剩余部分     

// 取模结果的符号与除数相同，取模更常用于循环索引等场景
int mod_result = ((-7 % 4) + 4) % 4; // 结果是 1
printf("mod_result = %d\n", mod_result);// 表示在循环（模数空间）中的位置

模数空间特性：
