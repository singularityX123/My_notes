简单选择排序

#include <stdio.h>
//简单选择排序
void selectionSort(int arr[], int n) {
    for (int i = 0; i < n-1; i++) {
        int min_idx = i;
        for (int j = i+1; j < n; j++) {
            if (arr[j] < arr[min_idx]) {
                min_idx = j;
            }
        }

        int temp = arr[i];
        arr[i] = arr[min_idx];
        arr[min_idx] = temp;
    }
}

{

    return 0;
}

int num = 0b1011; 
int N = 2; //对第三位二进制位变化

int set1faction(int num, int N){
    num |= (1 << N);
    printf("num = %#b\n", num);

    return num;
}

int set0faction(int num, int N){
    num &= ~(1 << N);
    printf("num = %#b\n", num);

    return num;
}
