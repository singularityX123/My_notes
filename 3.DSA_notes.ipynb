# 线性表

定义

```c++17
线性表是包含若干元素的一个线性序列
记为:L = (a0, a1, ..., a(n-1))
其中 L是表名，n是表长，ai (0<=i<=n-1)是表中的元素
             n > 0 时称为非空表，n = 0 时称为空表。

L = (D,R)  
D是数据元素的集合 R是数据元素之间的关系集合

/*从纯理论的数据结构角度来看，一个线性表（如数组、链表）通常被定义为一组相同类型的数据元素的有限序列。这种“类型相同”的约束是为了：
操作一致性：确保对每个元素进行的操作（如比较、排序、计算）都是有效的。
内存管理：相同类型的元素占用相同大小的内存空间，使得通过索引进行常数时间访问（例如在数组中）成为可能。
类型安全：避免在编译时或运行时出现类型错误。因此，在《数据结构》教科书中，通常默认线性表中的元素类型是相同的。*/
```

排序

```c++17
/*时间复杂度  空间复杂度  是否稳定*/

// 简单选择排序  O(n^2)  O(1)  不稳定    
// 每次从未排序部分选择最小元素放到已排序部分的末尾
void selectionSort(int arr[], int n) {
    for (int i = 0; i < n-1; i++) {
        int min_idx = i;
        for (int j = i+1; j < n; j++) {
            if (arr[j] < arr[min_idx]) {
                min_idx = j;
            }
        }
        // 交换
        int temp = arr[i];
        arr[i] = arr[min_idx];
        arr[min_idx] = temp;
    }
}

// 冒泡排序  O(n^2)  O(1)  稳定    
// 重复遍历数组，每次比较相邻元素，如果顺序错误就交换
void bubbleSort(int arr[], int n) {
    for (int i = 0; i < n-1; i++) {
        for (int j = 0; j < n-i-1; j++) {
            if (arr[j] > arr[j+1]) {
                // 交换
                int temp = arr[j];
                arr[j] = arr[j+1];
                arr[j+1] = temp;
            }
        }
    }
}
```
